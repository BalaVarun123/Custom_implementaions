```python
def predict(v,W,b):
    return W.dot(v) + b
def predict_all(V,W,b):
    P = []
    i = 0
    l = len(V)
    while (i<l):
        P.append(predict(V[i],W,b))
        i = i + 1
    return P
def cost(y_p,y_r):
    t = y_r*y_p
    if t >= 1:
        return 0
    else :
        return 1 - t
def total_cost(Y_p,Y_r):
    i = 0
    l = len(Y_r)
    summation = 0
    while (i<l):
        summation +=  cost(Y_p[i],Y_r[i])
        i  = i  + 1
    return summation
def fit(X,Y,W,b,learning_rate,regularizer,epochs,threshold = 0,output_when=100):
    iteration = 0
    new_W = W.copy()
    new_b = b
    D = len(W)
    P = predict_all(X,W,b)
    E = total_cost(P,Y)
    print(E)
    epoch = 0
    l = len(Y)
    while(E>threshold and iteration < epochs):
        i = 0
        while (i<l):
            d = 0
            #print("ffff")
            while (d < D):
                p = predict(X[i],new_W,new_b)
                if p*Y[i] < 1:
                    new_W[d] += learning_rate*(Y[i]*X[i][d] - 2*regularizer*new_W[d])
                else :
                    new_W[d] += learning_rate*(-2*regularizer*new_W[d])
                d += 1
            p = predict(X[i],W,b)
            if p*Y[i] < 1:
                    new_b += learning_rate*(-Y[i])
            i = i + 1
            #print(i)
        iteration += 1
        P = predict_all(X,new_W,new_b)
        E = total_cost(P,Y)
        if iteration % output_when == 0:
            print("Iteration:%d"%iteration)
            print(new_W,new_b)
            print("Total cost:%f"%E)
    return new_W,new_b
```

```python
X = []
Y = []
for i in range(150):
    l = []
    l.append(data["SepalLengthCm"][i])
    l.append(data["SepalWidthCm"][i])
    l.append(data["PetalLengthCm"][i])
    l.append(data["PetalWidthCm"][i])
    X.append(l)
    if data["Species"][i] == 'Iris-setosa':
        Y.append(-1)
    else :
        Y.append(1)
X = np.array(X)
Y = np.array(Y)
```
