```python
import numpy as np
import matplotlib.pyplot as plt
import math
#from decimal import Decimal
```

```python
DX = 1e-2
```

```python
def function(x):
    return -1*math.sin(x)
def func_arr(func,arr):
    return [func(i) for i in arr]
def nthderivative(f,n,x,dx = DX):
    if n == 0:
        return f(x)
    else :
        dy = nthderivative(f,n-1,x+dx,dx) - nthderivative(f,n-1,x,dx)
        return dy/dx
def taylor(f,n,x,a = 0, dx = DX):
    summation = 0
    for i in range(n):
        summation += (nthderivative(f,i,x,dx) / math.factorial(i)) * math.pow(x-a,i)
    return summation
```

```python
X = np.arange(-10,10,0.1)
Y = [nthderivative(function,2,x) for x in X]
plt.plot(X,Y)
Z = np.sin(X)
plt.plot(X,Z)
```
